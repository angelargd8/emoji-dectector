```python
import os
import shutil
```

```python

image_root = "./archive/image"
output_root = "./faces"

# Diccionario de clasificaciones con sus índices
classifications = {
    "enojado": [92, 93, 94, 95],
    "feliz": [1, 2, 3, 4, 5, 9],
    "muerto": [59, 98, 99],
    "confundido": [60],
    "neutral": [35, 36, 37],
    "sorprendida": [72, 73, 74, 75, 77, 78, 79, 84],
    "triste": [22, 68, 70, 81, 82, 83, 87, 88]
}

# Crear las carpetas de salida si no existen
os.makedirs(output_root, exist_ok=True)
for cls in classifications:
    os.makedirs(os.path.join(output_root, cls), exist_ok=True)

# Recorrer todas las carpetas (Apple, Google, etc.)
for brand in os.listdir(image_root):
    brand_path = os.path.join(image_root, brand)
    if not os.path.isdir(brand_path):
        continue

    # Procesar cada clasificación
    for cls, indices in classifications.items():
        for idx in indices:
            filename = f"{idx}.png"
            src = os.path.join(brand_path, filename)
            if os.path.exists(src):
                dst = os.path.join(output_root, cls, f"{brand}_{filename}")
                shutil.copy(src, dst)
                print(f"Copiado: {src} -> {dst}")
            else:
                print(f"No encontrado: {src}")

print("\nClasificación completada.")




```
